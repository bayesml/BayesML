```python
from bayesml import metatree,linearregression,normal,categorical
import numpy as np
```

```python
rng = np.random.default_rng(12345)
n = 100
x_continuous = rng.random((n,5))
x_categorical = rng.choice([0,1], size=(n,5))
y_continuous = rng.random(n)
y_categorical = rng.choice([0,1], size=n)
print(x_continuous)
print(x_categorical)
print(y_continuous)
print(y_categorical)
```

```python
model = metatree.LearnModel(
    5,
    5,
    2,
    2,
    SubModel=linearregression,
    sub_constants={'c_degree':5},
    sub_h0_params={'h0_alpha':1.1},
)
```

```python
model.update_posterior(
    x_continuous=x_continuous,
    x_categorical=x_categorical,
    y=y_continuous,
)
```

```python
model.calc_pred_dist(
    x_continuous=x_continuous,
    x_categorical=x_categorical
)
```

```python
model.make_prediction()
```

```python
model.calc_pred_var()
```

```python
model.calc_pred_density(np.arange(3)[np.newaxis,:])
```
